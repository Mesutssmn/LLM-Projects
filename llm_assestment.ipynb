```python
import torch
from transformers import TFGPT2LMHeadModel, GPT2Tokenizer
```

```python
model_name = "gpt2-large"
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = TFGPT2LMHeadModel.from_pretrained(model_name, pad_token_id=tokenizer.eos_token_id)
```

```python
def generate_code(input_index, max_length=100):
  input_text = "Input: " + dataset["train"][input_index]["instruction"] + "\nOutput:"

  # Encode input text into ids- tokenization
  input_ids = tokenizer.encode(input_text, return_tensors="tf")

  # Generate text
  output = model.generate(input_ids, max_length=max_length)

  # decode the ids back into text
  generated_lst = []
  for sample in output:
    generated_codes = tokenizer.decode(sample, skip_special_tokens=True)
    generated_lst.append(generated_codes)

  return generated_lst
```

```python
generate_code(0, max_length=250)
```

```python
generate_code(2, max_length=500)
```

```python
generate_code(3, max_length=250)
```

## Triyng Another Func.

```python
def generate_code_from_index(data_index, max_length=100):
    data_input = dataset["train"][data_index]["instruction"]

    # Tokenize the input instruction
    input_ids = tokenizer.encode(data_input, return_tensors="pt", max_length=max_length, truncation=True)

    # Generate output based on tokenized input
    output = model.generate(input_ids, max_length=max_length)

    generated_ads = []
    for sample in output:
        generated_output = tokenizer.decode(sample, skip_special_tokens=True)
        generated_ads.append(generated_output)

    print("############\n",Fore.RED +"Input Data :\n"+ Style.RESET_ALL,"############",data_input, "############",Fore.YELLOW + "\nGenerated Data :\n","############"+ Style.RESET_ALL, generated_ads, "############",Fore.BLUE + "\nOutput Data :\n","############"+ Style.RESET_ALL, dataset["train"][data_index]["output"], sep="\n")
```

```python
generate_code_from_index(0, max_length=250)
```

```python
import os
from getpass import getpass
import openai


openai_key = getpass("Enter your OpenAI API key:")
os.environ['OPENAI_API_KEY'] = openai_key
openai.api_key = os.environ['OPENAI_API_KEY']
```

```python
from langchain.document_loaders.csv_loader import CSVLoader #source+load
from langchain.embeddings.openai import OpenAIEmbeddings #transform+embed
from langchain.vectorstores import FAISS #vector database/stor
from langchain.chains import ConversationalRetrievalChain #retrival
from langchain.chat_models import ChatOpenAI #chatGpt model
from langchain.memory import ConversationBufferMemory #memory
```

```python
loader = CSVLoader(file_path="Cities_ChatGPT.csv")
data = loader.load()
```

```python
embeddings = OpenAIEmbeddings()
```

```python
vectorstore = FAISS.from_documents(data, embeddings)
```

```python
chatbot = ChatOpenAI(temperature=0.0,model_name='gpt-3.5-turbo')
```

```python
memory = ConversationBufferMemory(memory_key="chat_history", return_messages=True)
```

```python
chain = ConversationalRetrievalChain.from_llm(
    llm = chatbot,
    retriever=vectorstore.as_retriever(),
    memory = memory)
```

```python
question1 = "What are the best cities to live in Netherlands?"
answer1 = chain({"question":question1})
print(answer1['answer'])
```

```python
question2 = "What are the best cities to live in Belgium?"
answer2 = chain({"question":question2})
print(answer2['answer'])
```

```python
question3 = "What are the best cities to live in Turkey?"
answer3 = chain({"question":question3})
print(answer3['answer'])
```
